**Model Architecture:**

#### Handling Class Imbalance (Oversampling)

To address the skewed class distribution, we experimented with several data augmentation and cleaning strategies:

1.  **Full Balancing:** We initially oversampled all minority classes to match the majority class (approx. 600 samples). This led to poor generalization and decreased performance, likely due to **overfitting** on synthetic noise within the smallest classes.
2.  **Thresholded Oversampling:** We adjusted the strategy to only upsample classes with fewer than 50 samples. While this yielded a marginal improvement in the **F1-score**, the change was not statistically significant.
3.  **Data Pruning:** We attempted to remove "noisy" data by dropping classes with fewer than 10 total instances. This was intended to stabilize cross-validation results. While this did not yield a measurable leap in metrics, qualitative visual analysis of the confusion matrices suggested a slightly cleaner classification boundary.

## Software planning

Use MediaPipe as our pose estimation library, this functionality is implemented in the frontend to off load computation to the client side to avoid overloading our free hosting services. Current state we can run the pose estimation on live webcam feed and also uploaded videos, will implement image upload estimation in the upcoming week. The pose estimation keypoints are sotred as JSONB in our PostgreSQL database hosted on Supabase. 

This is how we load it in the frontend to avoid bloating our installation bundle.
```javascript
async function loadPoseLandmarker() {
    const { PoseLandmarker, FilesetResolver } =
        await import("https://cdn.jsdelivr.net/npm/@mediapipe/tasks-vision@0.10.3/+esm");

    const vision = await FilesetResolver.forVisionTasks(
        "https://cdn.jsdelivr.net/npm/@mediapipe/tasks-vision@0.10.3/wasm",
    );

    return PoseLandmarker.createFromOptions(vision, {
        baseOptions: {
            modelAssetPath:
                "https://storage.googleapis.com/mediapipe-models/pose_landmarker/pose_landmarker_lite/float16/1/pose_landmarker_lite.task",
            delegate: "GPU",
        },
        runningMode: "VIDEO",
        numPoses: 1,
    });
}
```

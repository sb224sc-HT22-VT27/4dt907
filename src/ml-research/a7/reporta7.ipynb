Interpreting the clusters:

The clustering process reveals that movement failures are not random, instead they follow four specific "styles" of compensation as our K was set to 4. By examining the sensors with the highest deviations, we can categorize these into distinct profiles.

Cluster 1 represents the most significant form breakdown, with a high deviation average. This group is defined by errors in the right hip, right thigh, and left arm (Sensors 11, 12, and 5). This indicates a collapse in the center of the body. Because these users cannot stabilize their hips and thighs, their spine likely leans too far forward or shifts, forcing the arms out of their ideal path just to maintain balance.

Cluster 0 is the stiff group, primarily failing on back verticality and head position (Sensors 23 and 20). Unlike the first group, these users keep their arms and legs relatively steady, but they struggle with moving smoothly. They move as if they are rigid, likely tilting the head too far forward or failing to keep the back upright as they descend into the movement.

Cluster 2 contains the side movers. Their errors are concentrated in shoulder leveling and left hip position (Sensors 19 and 8). Rather than moving straight up and down, these individuals lean to one side.

Cluster 3 is defined by right arm reach and mid-back alignment (Sensors 7 and 2). These movers actually have fairly steady legs and hips, but they "cheat" for balance using their upper body.

### PCA


![ClusterPlot](img/Clustering.png)


Interpreting the PCA plot:

The Principal Component Analysis (PCA) takes all 38 sensor dimensions and squashes them into a 2D map to show how "distant" these users are from one another.

The first axis, PC1 (the X-axis), acts as a scale for Movement Severity. The top contributors to this axis are sensors 4, 10, 5, 13, and 12, which cover the shoulders, arms, and legs. Moving from left to right on the plot is effectively moving from "Perfect Form" to "Total Breakdown." This is why Cluster 1 is pushed all the way to the right side of the map as they are mathematically the furthest away from the ideal movement pattern.

The second axis, PC2 (the Y-axis), describes the Style of Error. This axis separates users based on how they are failing—for example, distinguishing someone who is leaning too far forward from someone who is shifting weight to the side.

Looking at the scatterplot as a whole, we see that most people (Clusters 0, 2, and 3) are bunched together on the left side of the plot. This suggests they share a common baseline of "average" movement habits. However, Cluster 1 sits as a lonely island on the far right. This confirms that when the hips and thighs fail together, it creates a unique signature that is different from the rest of the data, making them easy to identify as the primary group needing improvement.

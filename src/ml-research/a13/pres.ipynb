# A13

## Pipeline
1. **Cut sequences** — A12 start/stop model trims raw recordings to the active exercise window
2. **Augment** — 25 variants per original (mirror, rotate, scale, noise) ~2 0000 sequences
3. **Feature engineering** — adds 16 inter-joint distances + 6 joint angles per frame
4. **Resample** — every sequence reduced to c=10 equidistant frames
5. **Train** — 10-fold grouped CV (augmented copies always stay on the training side, test uses originals only)

## Models
| Model | Architecture | Params |
|-------|-------------|--------|
| ADense | FC → BN → ReLU → Dropout → FC → 1 | ~10 000 |
| ACNN | Conv1D → MaxPool → FC → 1 | ≤ 20% of ADense |

## Experiments tested
- Baseline Dense and CNN
- Deeper Dense (more hidden units, higher dropout)
- Slow learning rate with large batch
- CNN with larger kernel
- CNN without poolings

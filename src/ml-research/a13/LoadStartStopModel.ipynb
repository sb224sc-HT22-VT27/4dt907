```python
import mlflow
import torch
import dagshub
import os

# 1. Initialize DagsHub connection
dagshub.init(repo_owner="SamuelFredricBerg", repo_name="4dt907", mlflow=True)

# 2. Define the model URI using the @prod alias
model_name = "Start_Stop_Predictor_ModelV2"
model_uri = f"models:/{model_name}@prod"

# 3. Load the model directly into memory
# This automatically handles the versioning logic based on the alias
print(f"Fetching model version tagged as @prod for: {model_name}...")
prod_model = mlflow.pytorch.load_model(model_uri)

# 4. Save the weights to a local .pth file
output_filename = "best_start_stop_model_prod.pth"
torch.save(prod_model.state_dict(), output_filename)

print(f"Successfully downloaded and saved: {output_filename}")
```

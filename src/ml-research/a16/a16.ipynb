## (New) endpoint

`api/v1/squat/analyze-session` on the backend which is only accessible via the [website](https://4dt907.vercel.app/) due to allowed origins configuration.

This endpoint is not "new" due to early adoption, realizing that all the smaller endpoints were meant to be merged into one later on and thus have been keping it as a single endpoint but making additions to the internal pipeline for said endpoint call.

Passes all of our local unit tests and smoke-tests, enpoint call is deployed in the frontend meanwhile the backend implements the pipeline to pruduce expected result. The way to use the mentioned endpoint is by on the [website](https://4dt907.vercel.app/) navigating to the "Squat Analyzer" tab and then either record a video via your webcam, upload existing video or an image, which when processed will send data extracted for said input to the endpoint and then display the results on the frontend.

## Website usage & Explanation

![One](img/one.png)
This is the page you arrive at when entering or website, you can navigate to the "Squat Analyzer" tab to get to the page where you can use the new endpoint in the top left of the page.

![Two](img/two.png)
After reaching the "Squat Analyzer" tab, the preselected alternative is the option to record a video via your webcam but with the choice to upload an existing video or an image in a centered menu. Press the "Start Recording" button to start recording a video via your webcam, if camera quality is deemed not good enough warnings or errors will present themselves within the preview window. Once you finished recording your squat press the "Stop & Analyze" button to send the video to the backend for processing and then display the results on the frontend.

![Three](img/three.png)
If you choose to upload an existing video or an image, you can do so by either clicking the "Choose video/image" button or by dragging and dropping the file into the upload area, this is in the area of the button. Once the file is uploaded it will be displayed in the preview windows for keypoint analysis before being sent to the backend for processing, to later display the results on the frontend.

![Four](img/four.png)
After the video/image is processed first in the frontend for its keypoints and then by the backend, this is the view you will be presented with, to see the results of the analysis scroll down the page.

![Five](img/five.png)
Here it is displayed wether the video or image uploaded is deemed to be a Squat exercise or not, note that if the squat is badly performed it more then likely be classified as "Not a squat" due to the keypoints not matching the expected pattern of a squat. The user has the option to adjust the allowed thereshold for the classification by adjusting the slider, which will make the classification more or less strict, notice the bar above the slider showcases how likely it thinks to uploaded file contained a squat exercise. Below the result reading "A Squat" or "Not a squat" is the score of the squat performed which is a value bewtween 0.0 and 4.0, the higher the score the worse the squat was performed.

The number of recorded frames are also displayed along side the options to "Save to database", "Download CSV" and "New session".
The "Save to database" button will save the keypoints to our database for future training of machine learning models. The "Download CSV" button will download a CSV file containing the keypoints. The "New session" button will reset the page and allow the user to start a new session of recording or uploading a video/image for analysis.

At the bottom of the page we have a 3D visualization of the keypoints extracted from the video/image, this is a nice way for users to visually see the keypoints and how they move during the squat exercise. This skeletion can be manipulated by the user to view it from different angles and perspectives, this is a nice way to see the keypoints in a more interactive way. With a slider to showcase certain frames of the video.

![Six](img/six.png)
As a user might notice there is a checkbox to toggle the "Pipeline timing window", this is a window that shows the time taken for each step in the pipeline, this is not something all users might be interested in seeing and thus it is optional to view. This is a nice way for users to see how long each step in the pipeline takes and where potential bottlenecks might be.

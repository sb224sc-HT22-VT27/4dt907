# Assignment 11: Gatekeeper Implementation Summary

This project implements a temporal "Gatekeeper" model designed to automatically identify the start and end of exercise sequences from skeletal coordinate data.

## 1. Data Preparation & Augmentation
* **Mirroring:** To improve model generalization and double the dataset size, all coordinate files were mirrored across the y-axis. This involved flipping X-coordinates and swapping left/right joint labels.
* **Automated Labeling:** Ground truth binary labels (0 for Idle, 1 for Active) were generated by mapping the frame ranges from manually "cut" exercise files onto the original "uncut" sensor sequences.

## 2. Handling Data Imbalance
* **Class Weighting:** Since "Idle" frames significantly outnumber "Active" frames, the dataset is heavily imbalanced. We calculated a `pos_weight` ratio based on the total distribution of labels.
* **Weighted Loss:** This ratio was fed into the `BCEWithLogitsLoss` function to penalize the model more heavily for misclassifying the minority (Active) class, ensuring the model prioritizes finding the exercise.

## 3. Deep Learning Architecture
* **LSTM Network:** A Long Short-Term Memory (LSTM) network was used to process temporal patterns. Unlike a standard neural network, the LSTM maintains a "memory" of previous frames, which is essential for recognizing movement intent.
* **Sliding Window (Temporal Context):** Data was processed in sequences of 30 frames. This provides the model with velocity and direction context, allowing it to accurately distinguish between static poses and active exercise transitions.

## 4. Robust Evaluation (10-Fold Cross-Validation)
* **Subject-Level Splitting:** To ensure valid results, 10-fold cross-validation was performed by splitting data based on unique subjects. This prevents "data leakage" (the model cannot be tested on a subject it has already seen in training).
* **Early Stopping:** A patience-based monitor was implemented to stop training once the validation loss ceased to improve, preventing overfitting and ensuring the model remains generalizable.

## 5. Performance Metrics & Model Selection
* **F1-Score Optimization:** The "Optimal Setup" was determined by identifying the model with the highest F1-Score (the balance between Precision and Recall).
* **Automated Sequence Cutting:** The resulting "Absolute Best" model is used as a Gatekeeper to automatically extract frame indices for the "Software Development" task, identifying the exact moment an exercise begins and ends.
